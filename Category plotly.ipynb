```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.countplot(data=tips, x='day')
plt.title("Total Records per Day")
plt.show()
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.title("Records per Day by Gender")
plt.show()
```

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.title("Horizontal Countplot: Day vs Gender")
plt.show()
```

```python
sns.countplot(data=tips, x='day', palette='Set2')
plt.title("Colored Countplot: Tips by Day")
plt.show()
```

```python
sns.set_style('darkgrid')
sns.countplot(data=tips, x='sex', hue='smoker')
plt.title("Gender vs Smoking Count")
plt.show()
```

```python
sns.countplot(data=tips, x='time')
plt.title("Count of Lunch vs Dinner")
plt.xticks(rotation=45)
plt.show()
```
