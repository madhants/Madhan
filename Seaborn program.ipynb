```python

import seaborn as sns
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
math_scores = [85, 90, 78, 92, 88]

sns.barplot(x=students, y=math_scores)
plt.title("Math Scores - Seaborn Bar Plot")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
science_scores = [80, 85, 75, 95, 89]

sns.lineplot(x=students, y=science_scores, marker='o')
plt.title("Science Scores - Seaborn Line Plot")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
math_scores = [85, 90, 78, 92, 88]

plt.bar(students, math_scores)
plt.title("Math Scores - Matplotlib Bar Chart")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
science_scores = [80, 85, 75, 95, 89]

plt.plot(students, science_scores, marker='o', color='green')
plt.title("Science Scores - Matplotlib Line Chart")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
math_scores = [85, 90, 78, 92, 88]
science_scores = [80, 85, 75, 95, 89]

plt.plot(students, math_scores, marker='o', label='Math')
plt.plot(students, science_scores, marker='s', label='Science')
plt.title("Math and Science Scores")
plt.xlabel("Student")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
scores = [88, 75, 90, 85, 80]

plt.barh(students, scores, color='orange')
plt.title("Scores by Student - Horizontal Bar Chart")
plt.xlabel("Score")
plt.ylabel("Student")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

math_scores = [85, 90, 78, 92, 88]
science_scores = [80, 85, 75, 95, 89]

sns.scatterplot(x=math_scores, y=science_scores)
plt.title("Math vs Science Scores - Seaborn Scatter Plot")
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.show()
```
