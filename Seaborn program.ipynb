```python

import seaborn as sns
import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
math_scores = [85, 90, 78, 92, 88]

sns.barplot(x=students, y=math_scores)
plt.title("Math Scores - Seaborn Bar Plot")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```
