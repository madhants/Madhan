```python
import plotly.express as px
import pandas as pd

# Load the iris dataset
df = px.data.iris()

# Select only numeric columns (sepal and petal)
features = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# Calculate the correlation matrix
corr_matrix = features.corr()

# Plot using Plotly heatmap
fig = px.imshow(
    corr_matrix,
    text_auto=True,
    color_continuous_scale='Viridis',
    title="Correlation Matrix: Sepal & Petal Length/Width"
)

fig.show()
```

```python
fig = px.density_heatmap(
    df,
    x='sepal_length',
    y='petal_length',
    nbinsx=20,
    nbinsy=20,
    color_continuous_scale='Turbo',
    title="Density Heatmap: Sepal Length vs Petal Length"
)
fig.show()
```

```python
# Correlation of petal length vs petal width
petal_corr = df[['petal_length', 'petal_width']].corr()

fig = px.imshow(
    petal_corr,
    text_auto=True,
    color_continuous_scale='Inferno',
    title='Correlation Matrix: Petal Length vs Width'
)
fig.show()
```

```python
import plotly.express as px

df = px.data.iris()

fig = px.scatter(
    df,
    x="sepal_length",
    y="petal_length",
    color="species",
    title="Comparison: Sepal Length vs Petal Length by Species",
    symbol="species"
)
fig.show()
```

```python
fig = px.violin(
    df,
    x="species",
    y="sepal_width",
    color="species",
    box=True,
    points="all",
    title="Violin Plot: Sepal Width by Species"
)
fig.show()
```

```python
import pandas as pd

# Calculate average values
avg = df.groupby('species')[['sepal_length', 'petal_length']].mean().reset_index()

# Melt the dataframe to long format for Plotly
avg_melted = avg.melt(id_vars='species', var_name='Measurement', value_name='Length')

fig = px.bar(
    avg_melted,
    x='species',
    y='Length',
    color='Measurement',
    barmode='group',
    title="Average Sepal vs Petal Length by Species"
)
fig.show()
```

```python
fig = px.box(
    df,
    x="species",
    y="petal_width",
    color="species",
    title="Box Plot: Petal Width by Species"
)
fig.show()
```

```python
fig = px.scatter_matrix(
    df,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix: All Feature Comparison"
)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
df = sns.load_dataset("iris")

# Use only numeric features
features = df.drop("species", axis=1)

# Create clustermap
sns.clustermap(features, cmap="viridis", standard_scale=1)
plt.title("Clustermap of Iris Dataset Features")
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Sample wide-format data
data = {
    'Category': ['A', 'B', 'C'],
    'Score_1': [10, 15, 20],
    'Score_2': [30, 25, 35]
}
df = pd.DataFrame(data)

# Plot using wide format (each column as one bar series)
fig = px.bar(df, x='Category', y=['Score_1', 'Score_2'], barmode='group', title="Grouped Bar: Wide Format")
fig.show()
```

```python
# Map species to colors
lut = dict(zip(df['species'].unique(), ['red', 'green', 'blue']))
row_colors = df['species'].map(lut)

sns.clustermap(features, cmap="coolwarm", row_colors=row_colors, standard_scale=1)
plt.title("Clustermap with Species Color Coding")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Create regression plot
sns.regplot(data=tips, x='total_bill', y='tip', color='blue')

plt.title("Regression Plot: Tip vs Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
import plotly.express as px
import numpy as np

df = px.data.tips()

# Fit regression line using NumPy
coeffs = np.polyfit(df['total_bill'], df['tip'], deg=1)
slope, intercept = coeffs

# Add predicted values
df['predicted_tip'] = slope * df['total_bill'] + intercept

# Scatter plot with regression line
fig = px.scatter(df, x='total_bill', y='tip', color='sex', title="Regression (Plotly + NumPy)")
fig.add_scatter(x=df['total_bill'], y=df['predicted_tip'], mode='lines', name='Regression Line')
fig.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title("Linear Regression by Gender")
plt.show()
```
